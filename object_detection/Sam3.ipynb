```python
from ultralytics.models.sam import SAM3SemanticPredictor

# Initialize predictor with configuration
overrides = dict(
    conf=0.25,
    iou=0.7,
    task="segment",
    mode="predict",
    model="/home/jupyter/project/research/object_detection/sam3.pt",
    half=True,  # Use FP16 for faster inference
    save=True,
    save_crop=True
)
predictor = SAM3SemanticPredictor(overrides=overrides)
```

```python
predictor.set_image("/home/jupyter/project/research/experiments/table.jpeg")

results = predictor(text=[
    "calendar",
    "highlighter",
    "desk lamp",
    "notebook",
    "pen",
    "mouse",
    "phone",
    "fan",
    "water bottle",
    "charger"
    ])
```
